#Создадим кастомный Trainer для дистилляции знаний

```python
import os
import time

import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F

from datasets import load_dataset, load_metric
from transformers import TrainingArguments, Trainer, AutoTokenizer, AutoConfig, AutoModelForSequenceClassification, pipeline
```

```python
class KnowledgeDistillationTrainingArguments(TrainingArguments):
  def __init__(self, *args, alpha=0.5, temperature=2.0, **kwargs):
    super().__init__(*args, **kwargs)

    self.alpha = alpha
    self.temperature = temperature
```

#Напишем лосс-функцию для дистилляции знаний

Создадим наследника класса Trainer и перепишем compute_loss()

```python
class KnowledgeDistillationTrainer(Trainer):
  def __init__(self, *args, teacher_model=None, **kwargs):
    super().__init__(*args, **kwargs)
    self.teacher_model = teacher_model

  def compute_loss(self, model, inputs, return_outputs=False):
    #Extract cross-entropy loss and logits from student
    outputs_student = model(**inputs)
    loss_ce = outputs_student.loss
    logits_student = outputs_student.logits

    # Extract logits from teacher
    outputs_teacher = self.teacher_model(**inputs)
    logits_teacher = outputs_teacher.logits

    #Computing distillation loss by Softening probabilities
    loss_fct = nn.KLDivLoss(reduction="batchmean")
    #The reduction=batchmean argument in nn.KLDivLoss() specifies that we average the losses over the batch dimension.
    loss_kd = self.args.temperature ** 2 * loss_fct(
                F.log_softmax(logits_student / self.args.temperature, dim=-1),
                F.softmax(logits_teacher / self.args.temperature, dim=-1))

    # Return weighted student loss
    loss = self.args.alpha * loss_ce + (1. - self.args.alpha) * loss_kd
    return (loss, outputs_student) if return_outputs else loss
```

#Загрузка датасета

Загрузим датасет tweet_sentiment_extraction

```python
dataset = load_dataset("mteb/tweet_sentiment_extraction")
```

Датасет содержит тексты, а также их эмоциональную окраску

```python
sample = dataset["train"][0]
print(sample)
```

Разделим и токенизируем датасет

```python
def tokenize_batch(batch):
  return student_tokenizer(batch["text"], return_tensors="pt", truncation=True, padding=True,
                              max_length=50)
```

```python
train_dataset, val_dataset = dataset['train'].train_test_split(test_size=0.1).values()
```

```python
tokenized_train = train_dataset.map(tokenize_batch, batched=True, remove_columns=["id","text",'label_text'])
```

```python
tokenized_val = val_dataset.map(tokenize_batch, batched=True, remove_columns=["id","text",'label_text'])
```

```python
tokenized_val = tokenized_val.rename_column("label","labels")
```

```python
tokenized_train = tokenized_train.rename_column("label","labels")
```

# Определим метрику, которой будем замерять точность

```python
accuracy_score = load_metric("accuracy")

def compute_metrics(pred):
  predictions, labels = pred
  predictions = np.argmax(predictions, axis=1)
  return accuracy_score.compute(predictions=predictions, references=labels)
```


# Определим аргументы для тренировки

```python
batch_size = 48
finetuned_student_ckpt = "bert-tiny-finetuned-tse-student"
```

```python
student_training_args = KnowledgeDistillationTrainingArguments(
    output_dir=finetuned_student_ckpt, evaluation_strategy = "epoch",
    num_train_epochs=1, learning_rate=1e-5,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size, alpha=1, weight_decay=1e-3)
```

# Инициализируем модель-ученика

```python
label2id = {"positive": 2, "neutral": 1, "negative": 0}
id2label = dict(zip(label2id.values(),label2id.keys()))
```

```python
num_labels = len(label2id)
student_config = (AutoConfig.from_pretrained(student_id,
                                             num_labels=num_labels,
                                             id2label=id2label,
                                             label2id=label2id))
```

```python
def student_init():
  return (AutoModelForSequenceClassification.from_pretrained(student_id, config=student_config).to(device))
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

# Загрузим обученную ранее модель-учителя и начнем дообучение модели-ученика

```python
teacher_model = torch.load('/content/model-compression-course/baseline.pt', map_location=device)
```

```python
# Начинаем fine-tuning ученика
distilbert_trainer = KnowledgeDistillationTrainer(model_init=student_init,
        teacher_model=teacher_model, args=student_training_args,
        train_dataset=tokenized_train, eval_dataset=tokenized_val,
        compute_metrics=compute_metrics, tokenizer=student_tokenizer)

distilbert_trainer.train()
```

```python
teacher_model.save_pretrained("teacher_model")
distilbert_trainer.save_model('student_model')

raw_student = AutoModelForSequenceClassification.from_pretrained(student_id, config=student_config)
raw_student.save_pretrained("raw_student_model")
```
